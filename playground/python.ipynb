```python
from rdflib import Graph, URIRef, Literal, Namespace
from rdflib.namespace import RDF,RDFS
```

```python
g = Graph()
SE = Namespace("https://github.com/usalu/semio/releases/download/v1/semio-v1.ttl#")
MB = Namespace("https://github.com/usalu/semio/examples/metabolism")
g.bind("se",SE)
g.bind("mb",MB)
g.add((MB.prop1,SE.value,Literal("crv1",datatype=SE.curve)))
print(g.serialize())
```

```python
g = Graph()
SE = Namespace("https://github.com/usalu/semio/releases/download/v1/semio-v1.ttl#")
g.bind("se",SE)
g.add((URIRef("prop1"),SE.value,Literal("crv1",datatype=SE.curve)))
MB = Namespace("https://github.com/usalu/semio/examples/metabolism")
g.base = MB
print(g.serialize())
```
